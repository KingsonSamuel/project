```python
# Handle missing values
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
```

```python
# Handle missing values for 'Age' column
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].median())
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())

print(train_data['Age'].isnull().sum())  # Should print 0 if all missing values are filled
print(test_data['Age'].isnull().sum())  # Should print 0 if all missing values are filled
```

```python
# Encode categorical variables
train_data = pd.get_dummies(train_data, columns=['Sex', 'Embarked'])
test_data = pd.get_dummies(test_data, columns=['Sex', 'Embarked'])
```

```python
# Scale numerical features (optional but can be beneficial for certain algorithms)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_data[['Age', 'Fare']] = scaler.fit_transform(train_data[['Age', 'Fare']])
test_data[['Age', 'Fare']] = scaler.transform(test_data[['Age', 'Fare']])
```

```python
X_train = train_data.drop(['Survived', 'Name', 'Ticket', 'Cabin', 'PassengerId'], axis=1)
y_train = train_data['Survived']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_val)
accuracy = accuracy_score(y_val, y_pred)
print("Accuracy:", accuracy)
```

```python
X_test = test_data.drop(['Name', 'Ticket', 'Cabin', 'PassengerId'], axis=1)
predictions = model.predict(X_test)
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer object
imputer = SimpleImputer(strategy='median')

# Fit the imputer to your training data and transform both training and test data
train_data_imputed = imputer.fit_transform(train_data)
test_data_imputed = imputer.transform(test_data)

# Proceed with model prediction using the imputed test data
predictions = model.predict(test_data_imputed)
```
